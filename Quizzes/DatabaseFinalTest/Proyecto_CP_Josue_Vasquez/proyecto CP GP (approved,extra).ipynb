```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
link1 = "Datasets/student-mat.csv"
link2 = "Datasets/student-por.csv"
```

```python
school1 = pd.read_csv(link1, sep=';', usecols=['school', 'sex', 'age', 'address', 'Pstatus', 'guardian', 'traveltime', 'studytime', 'failures', 'paid', 'internet', 'health', 'absences', 'G1', 'G2', 'G3'])
school1.head(10)
```

```python
school2 = pd.read_csv(link2, sep=';', usecols=['school', 'sex', 'age', 'address', 'Pstatus', 'guardian', 'traveltime', 'studytime', 'failures', 'paid', 'internet', 'health', 'absences', 'G1', 'G2', 'G3'])
school2.head(10)
```

```python
school1=school1[school1!='MS']
school1=school1[school1['school'].notna()]
```

```python
approved=[]
extra=[]
aux=[]
```

```python
def puntaje1(valor):
        if  (93-valor*100/93<=80):
            return 0
        elif (93-valor*100/93 >80):
            return 1
```

```python
def puntaje2(valor):
        if  (valor<10):
            return 0
        elif (10<=valor<15):
            return 1
        elif (valor>=15):
            return 1
```

```python
def puntaje3(valor):
        if (10<=valor<=15):
            return 1
        elif (valor>=15):
            return 0
        elif (valor<10):
            return ""
```

```python
for valor in school1["absences"]:
    approved.append(puntaje1(valor))
```

```python
for valor in school1["G3"]:
    aux.append(puntaje2(valor))
```

```python
for x in range (len(approved)):
    if(approved[x]!=aux[x]):
        approved[x]=aux[x]
```

```python
for valor in school1["G3"]:
    extra.append(puntaje3(valor))
```

```python
school1["approved"]=approved
school1["extra"]=extra
```
